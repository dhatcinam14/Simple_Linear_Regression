#  Machine Learning

###  Employee Salary Prediction using Their Years of Experience

### Import Necessary Libraries like Pandas

```python
#to import pandas
import pandas as pd
```

```python
#to import salary data use pd.read_csv as dataset
dataset = pd.read_csv("Salary_Data.csv")
```

```python
#print dataset
dataset
```

### Split Input / Output

```python
#yearsexperience is dataset as independent
Independent = dataset[["YearsExperience"]]
```

```python
#print independent
Independent
```

```python
#salary is dataset as dependent
Dependent = dataset[["Salary"]]
```

```python
#print dependent
Dependent
```

```python
#import scikit-learn as sk
from sklearn.model_selection import train_test_split
```

```python
#train_test_split as x,y
X_train,X_test,y_train,y_test = train_test_split(Independent,Dependent,test_size=0.30,random_state=0)
```

```python
#print x_train
X_train
```

```python
#print y_train
y_train
```

```python
#print y_test
y_test
```

```python
#print x_test
X_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor
```

```python
regressor.fit(X_train,y_train)
```

```python
weight = regressor.coef_
```

```python
weight
```

```python
bias = regressor.intercept_
```

```python
bias
```

```python
Y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r_score = r2_score(y_test,Y_pred)
```

## r_score[near by 1 it is good mode

```python
r_score 
```
